```python
#Extract Only traffic_volume(Core Model)
traffic_data = df[['traffic_volume']]
traffic_data.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))
plt.plot(traffic_data, color='blue')
plt.title("Traffic Volume Over Time")
plt.xlabel("Date")
plt.ylabel("Traffic Volume")
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
```

```python
scaled_data = scaler.fit_transform(traffic_data)
```

```python
scaled_data[:5]
```

```python
scaled_data = scaler.fit_transform(traffic_data)
```

```python
def create_sequences(data, window_size=24):
    X = []
    y = []
    
    for i in range(len(data) - window_size):
        X.append(data[i:i+window_size])
        y.append(data[i+window_size])
        
    return np.array(X), np.array(y)
```

```python
X, y = create_sequences(scaled_data, 24)
```

```python
X.shape, y.shape
```

```python
#Train Test Split
split = int(0.8 * len(X))

X_train = X[:split]
X_test = X[split:]

y_train = y[:split]
y_test = y[split:]
```

```python
#Check Shapes
X_train.shape, X_test.shape
```

```python
#Build the LSTM Model
model = Sequential()

model.add(LSTM(50, return_sequences=True, input_shape=(24,1)))
model.add(LSTM(50))

model.add(Dense(1))

model.compile(
    optimizer='adam',
    loss='mean_squared_error'
)

model.summary()
```

```python
#Train the LSTM Model
history = model.fit(
    X_train,
    y_train,
    epochs=25,
    batch_size=32,
    validation_data=(X_test, y_test)
)
```

```python
#Traffic Prediction
predictions = model.predict(X_test)
```

```python
#Convert Predictions Back to Original Scale
predictions = scaler.inverse_transform(predictions)
actual = scaler.inverse_transform(y_test)
```

```python
#Calculate Model Accuracy
rmse = np.sqrt(mean_squared_error(actual, predictions))
mae = mean_absolute_error(actual, predictions)

print("RMSE:", rmse)
print("MAE:", mae)
```

```python
plt.figure(figsize=(12,6))

plt.plot(actual[:500], label="Actual Traffic")
plt.plot(predictions[:500], label="Predicted Traffic")

plt.title("Traffic Volume Prediction using LSTM")
plt.xlabel("Time")
plt.ylabel("Traffic Volume")
plt.legend()

plt.show()
```
